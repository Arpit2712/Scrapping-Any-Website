```python
import bs4
import requests
import re
```

```python
url=input("Enter Your Link:")
data=requests.get(url)

soup=bs4.BeautifulSoup(data.text,'html.parser')

print("---------------------------------------------------------")
print("**************** WELCOME TO WEB SCRAPPING****************")
print("---------------------------------------------------------")

# For Paragraph

def paragraph():
    lap=open("Para.txt","w")
    for para in soup.findAll('p'):
        lap.write('\n')
        lap.write(para.text)
    print("\nParagraph file created")
    lap.close()

# For Heading

def heading():
    while(True):
        ch=input("Enter heading Type(h1/h2/h3/h4):")
        if ch=="h1":
            print("You have entered h1-type")
            new=open("heading-h1.txt","w")
            for head in soup.findAll(ch):
                new.write('\n')
                new.write(head.text)
            print("\nHeading-h1 file created")
            new.close()
    
        elif ch=="h2":
            print("\nYou have entered h2-type")
            new=open("heading-h2.txt","w")
            for head in soup.findAll(ch):
                new.write('\n')
                new.write(head.text)
            print("\nHeading-h2 file created")
            new.close()

        elif ch=="h3":
            print("\nYou have entered h3-type")
            new=open("heading-h3.txt","w")
            for head in soup.findAll(ch):
                new.write('\n')
                new.write(head.text)
            print("\nHeading-h3 file created")
            new.close()
        
        elif ch=="h4":
            print("\nYou have entered h4-type")
            new=open("heading-h4.txt","w")
            for head in soup.findAll(ch):
                new.write('\n')
                new.write(head.text)
            print("\nHeading-h4 file created")
            new.close()
    
        else:
            print("\nYou have Entered wrong choice")
            a=input("\nDo you want to continue this loop(y/n):")
            if a=="y":
                continue
            else:
                break

# For Web_Links

def weblink():
    top=open("Weblink.txt","w")
    for link in soup.find_all('a',attrs={'href':re.compile("^https://")}):
        weblink=link.get('href')
        top.write('\n')
        top.write(weblink)
    top.close()
    print("\nWeblink File Created")

# For Image_Link

def image():
    mob=open("Image.txt","w")
    for image in soup.find_all('img'):
        image.append(image.get('src'))
        mob.write('\n')
        mob.write(image.text)
    mob.close()
    print("\nImage File Created")

# For Creating loop

while(True):
    print("\nPlease choose any options:")
    print("(1)Scrape paragraph")
    print("(2)Scrape Headine")
    print("(3)Scrape Weblink")
    print("(4)Scrape ImageLink")
    print("(5)Exit")
    ch=int(input("Enter Your Choice:"))
    
    if ch==1:
        paragraph()
    elif ch==2:
        heading()
    elif ch==3:
        weblink()
    elif ch==4:
        image()
    elif ch==5:
        break
    else:
        print("\nYou Entered Wrong Choice")
        
```
